# Summer Olympics Data Analysis Assignment

```python
import pandas as pd
import matplotlib.pyplot as plt

Data=pd.read_csv("summer.csv")
print("******************Dataset ********************")
print(Data.head())
```

### 1. In how many cities Summer Olympics is held so far?

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
x=Data["Sport"].unique()
for i in x :
    
    Data1=Data[Data["Sport"]==i]
    Data1=Data[Data["Medal"]=="Gold"]
lis2=[]
for i in Data1["Sport"].unique():
    lis2.append([i,len(Data1[Data1["Sport"] == i])])
lis2=pd.DataFrame(lis2,columns=["Sport","Number"])
lis2=lis2.sort_values(by="Number",ascending=False).head()
lis2.plot(x="Sport",y="Number",kind="bar")
plt.show()
```
